```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.ticker import FuncFormatter

sns.set()
sns.set_style("whitegrid")

out_file_path = "../img/"
```

```python
double_data = pd.read_csv("../data/dgemm-magma-4090.csv",
                      header=None,
                      names=[
                          "m", "n", "k", "mag_gflops", "mag_time", "cu_gflops", "cu_time", "mag_err"
                      ])

double_data=double_data.loc[:,["m", "mag_gflops","cu_gflops"]]

print(double_data)
```

```python
float_data = pd.read_csv("../data/sgemm-magma-4090.csv",
                      header=None,
                      names=[
                          "m", "n", "k", "mag_gflops", "mag_time", "cu_gflops", "cu_time", "mag_err"
                      ])

float_data=float_data.loc[:,["m", "mag_gflops","cu_gflops"]]

print(float_data)
```

```python
half_data = pd.read_csv("../data/hgemm-magma-4090.csv",
                      header=None,
                      names=[
                          "m", "n", "k", "mag_gflops", "mag_time"
                      ])

half_data=half_data.loc[:,["m", "mag_gflops"]]

print(half_data)
```

```python
notc_half_data = pd.read_csv("../data/hgemm-magma-4090-notc.csv",
                      header=None,
                      names=[
                          "m", "n", "k", "mag_gflops", "mag_time"
                      ])

notc_half_data=notc_half_data.loc[:,["m", "mag_gflops"]]

print(notc_half_data)
```

```python
notc_float_data = pd.read_csv("../data/sgemm-magma-4090-notc.csv",
                      header=None,
                      names=[
                          "m", "n", "k", "mag_gflops", "mag_time", "cu_gflops", "cu_time", "mag_err"
                      ])

notc_float_data=notc_float_data.loc[:,["m", "mag_gflops","cu_gflops"]]

print(notc_float_data)
```

```python
getrf_double_data = pd.read_csv("../data/dgetrf-magma-4090.csv",
                      header=None,
                      names=[
                          "bits", "pro", "m", "mag_gflops", "mag_time"
                      ])

getrf_double_data=getrf_double_data.loc[:,["m", "mag_gflops","mag_time"]]

print(getrf_double_data)
```

```python
getrf_float_data = pd.read_csv("../data/sgetrf-magma-4090.csv",
                      header=None,
                      names=[
                          "bits", "pro", "m", "mag_gflops", "mag_time"
                      ])

getrf_float_data=getrf_float_data.loc[:,["m", "mag_gflops","mag_time"]]

print(getrf_float_data)
```

```python
getrf_half_data = pd.read_csv("../data/hgetrf-magma-4090.csv",
                      header=None,
                      names=[
                          "bits", "pro", "m", "mag_gflops", "mag_time"
                      ])

getrf_half_data=getrf_half_data.loc[:,["m", "mag_gflops", "mag_time"]]

print(getrf_half_data)
```

```python
plt.figure(figsize=(7, 2.5), dpi=400)
plt.subplots_adjust(wspace=0.27, hspace=0.05)

dot_size = 2
legend_font_size = 10

plt.subplot(121)

plt.scatter((double_data["m"]),
            (double_data["cu_gflops"])/1e3,
            label="FP64",
            s=dot_size,
            )
plt.scatter((notc_float_data["m"]),
            (notc_float_data["cu_gflops"])/1e3,
            label="FP32",
            s=dot_size,
            )
plt.scatter((notc_half_data["m"]),
            (notc_half_data["mag_gflops"])/1e3,
            label="FP16",
            s=dot_size,
            )
plt.scatter((half_data["m"]),
            (half_data["mag_gflops"])/1e3,
            label="FP16 (TC)",
            s=dot_size,
            )

plt.legend(loc=2, fontsize=legend_font_size, markerscale=3.6)

plt.title("General Matrix Multiply")
plt.ylabel("TFLOPS on 4090")
plt.xlabel("Matirx Order (64-16384, step 64)")

plt.subplot(122)

plt.scatter((getrf_double_data["m"]),
            (getrf_double_data["mag_gflops"])/1e3,
            label="FP64",
            s=dot_size,
            )
plt.scatter((getrf_float_data["m"]),
            (getrf_float_data["mag_gflops"])/1e3,
            label="FP32",
            s=dot_size,
            )
plt.scatter((getrf_half_data["m"]),
            (getrf_half_data["mag_gflops"])/1e3,
            label="FP16",
            s=dot_size,
            )
# plt.scatter((getrf_half_data["m"]),
#             (getrf_half_data["mag_tc_gflops"])/1e3,
#             label="MAGMA FP16 (TC)",
#             s=dot_size,
#             )

plt.legend(loc=2, fontsize=legend_font_size, markerscale=3.6)

plt.title("LU Factorization")
plt.ylabel("TFLOPS on 4090")
plt.xlabel("Matirx Order (64-32768, step 64)")

plt.savefig ("gemm_and_getrf_4090.pdf", dpi= 400, format = "pdf", bbox_inches='tight', pad_inches=0.0) 
```
