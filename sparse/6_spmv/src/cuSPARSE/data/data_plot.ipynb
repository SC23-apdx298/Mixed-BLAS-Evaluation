```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

data = pd.read_csv("out.csv", header=None)

spmv_csr_16_16_32 = data.iloc[:77,:]
spmv_csr_16_32_32 = data.iloc[77:154,:]
spmv_csr_16BF_16BF_32 = data.iloc[154:231,:]
spmv_csr_16BF_32_32 = data.iloc[231:308,:]
spmv_csr_32F = data.iloc[308:385,:]
spmv_csr_64F = data.iloc[385:461,:]
spmv_csr_8_32_32 = data.iloc[461:538,:]
spmv_csr_8_32_32I = data.iloc[538:615,:]

print(spmv_csr_16_16_32)
print(spmv_csr_16_32_32)
print(spmv_csr_16BF_16BF_32)
print(spmv_csr_16BF_32_32)
print(spmv_csr_32F)
print(spmv_csr_64F)
print(spmv_csr_8_32_32)
print(spmv_csr_8_32_32I)
```

```python
plt.figure(figsize=(10,6), dpi=80)
plt.scatter(np.log(spmv_csr_16_16_32.iloc[:,7]), np.log(spmv_csr_16_16_32.iloc[:,8]), c='b', label="spmv_csr_16_16_32")
plt.scatter(np.log(spmv_csr_32F.iloc[:,7]), np.log(spmv_csr_32F.iloc[:,8]), c='g', label="spmv_csr_32F")
plt.scatter(np.log(spmv_csr_64F.iloc[:,7]), np.log(spmv_csr_64F.iloc[:,8]), c='r', label="spmv_csr_64F")
plt.legend(loc=0)
plt.xlabel("nnz")
plt.ylabel("Time")
plt.show()
```
